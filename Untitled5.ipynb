```python
def forecast_model(ts_train, steps=12, method='ets'):
    if not isinstance(ts_train.index, pd.DatetimeIndex):
        ts_train.index = pd.to_datetime(ts_train.index)
    ts_train = ts_train.resample('MS').sum()

    future_index = pd.date_range(
        start=ts_train.index[-1] + pd.DateOffset(months=1),
        periods=steps, freq='MS'
    )

    preds = None

    df = pd.DataFrame({'y': ts_train})
    df['trend'] = np.arange(len(df))
    df['month'] = ts_train.index.month
    X = pd.get_dummies(df[['trend', 'month']], drop_first=True)
    y = df['y'].values

    future_df = pd.DataFrame({
        'trend': np.arange(len(df), len(df) + steps),
        'month': future_index.month
    })
    Xf = pd.get_dummies(future_df, drop_first=True).reindex(columns=X.columns, fill_value=0)

    if method == 'stl_ets':
        stl_res = STL(ts_train, period=12).fit()
        tr, se = stl_res.trend, stl_res.seasonal
        fit_tr = ExponentialSmoothing(tr, trend='add').fit()
        fit_se = ExponentialSmoothing(se, seasonal='add', seasonal_periods=12).fit()
        preds = fit_tr.forecast(steps) + fit_se.forecast(steps)

    elif method == 'ets':
        model = ExponentialSmoothing(ts_train, trend='add', seasonal='add', seasonal_periods=12)
        fit = model.fit()
        preds = fit.forecast(steps)

    elif method == 'extra_trees':
        model = ExtraTreesRegressor()
        model.fit(X, y)
        preds = model.predict(Xf)

    else:
        raise ValueError(f"Unsupported method: {method}")

    return pd.Series(preds, index=future_index)
```

```python
def forecast_denver(ts_train, steps=12):
    future_index = pd.date_range(start=ts_train.index[-1] + pd.DateOffset(months=1), periods=steps, freq='MS')

    models = ['stl_ets']
    forecasts = {}

    for model_name in models:
        try:
            forecast = forecast_model(ts_train, steps, method=model_name)
            forecasts[model_name] = forecast
        except Exception as e:
            print(f"Model {model_name} failed: {e}")
            forecasts[model_name] = pd.Series([np.nan] * steps, index=future_index)

    return pd.DataFrame(forecasts, index=future_index)
```

```python
def forecast_hawaii(ts_train, steps=12):
    future_index = pd.date_range(start=ts_train.index[-1] + pd.DateOffset(months=1), periods=steps, freq='MS')

    models = ['stl_ets']
    forecasts = {}

    for model_name in models:
       try:
          forecast = forecast_model(ts_train, steps, method=model_name)
          forecasts[model_name] = forecast
       except Exception as e:
            print(f"Model {model_name} failed: {e}")
            forecasts[model_name] = pd.Series([np.nan] * steps, index=future_index)

    return pd.DataFrame(forecasts, index=future_index)
```

```python
def forecast_ashville(ts_train, steps=12):
    future_index = pd.date_range(start=ts_train.index[-1] + pd.DateOffset(months=1), periods=steps, freq='MS')

    models = ['extra_trees' ]
    forecasts = {}

    for model_name in models:
        try:
            forecast = forecast_model(ts_train, steps, method=model_name)
            forecasts[model_name] = forecast
        except Exception as e:
            print(f"Model {model_name} failed: {e}")
            forecasts[model_name] = pd.Series([np.nan] * steps, index=future_index)

    return pd.DataFrame(forecasts, index=future_index)
```

```python
def evaluate_forecasts(forecasts_df, actual_series):
    from sklearn.metrics import mean_absolute_error, mean_squared_error

    results = {}

    for model in forecasts_df.columns:
        forecast = forecasts_df[model]
        common_index = forecast.index.intersection(actual_series.index)

        if len(common_index) == 0:
            results[model] = {"MAE": None, "RMSE": None}
            continue

        forecast = forecast.loc[common_index]
        actual = actual_series.loc[common_index]

        mae = mean_absolute_error(actual, forecast)
        rmse = mean_squared_error(actual, forecast) ** 0.5
        results[model] = {"MAE": mae, "RMSE": rmse}

    return pd.DataFrame(results).T
```
